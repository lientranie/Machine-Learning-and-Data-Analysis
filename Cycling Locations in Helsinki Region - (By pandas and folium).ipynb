```python
import pandas as pd
import folium
```

```python
df_bike = pd.read_csv('hki.csv')
```

```python
# create a Stamen Toner map of the world centered around Helsinki
# folium.Map(location=[61.9241, 25.7482], zoom_start=4, tiles='Stamen Terrain')
latitude = 60.1699
longitude = 24.9384
Hel_map =  folium.Map(location=[latitude, longitude], zoom_start=12)

# instantiate a feature group for the incidents in the dataframe
bike = folium.map.FeatureGroup()

# loop through the data and add each to the incidents feature group
for lat, lng, in zip(df_bike.Y, df_bike.X):
    bike.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=6, # define how big the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        )
    )

# add pop-up text to each marker on the map
latitudes = list(df_bike.Y)
longitudes = list(df_bike.X)
labels = list(df_bike.Paikka)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(Hel_map)    
    
# add incidents to map
Hel_map.add_child(bike)
```
