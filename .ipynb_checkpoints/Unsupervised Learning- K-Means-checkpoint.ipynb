## Clustering

```python
from sklearn.datasets  import load_iris
from sklearn.cluster import KMeans
data = load_iris()
X = data.data
y = data.target
data.target_names
```

```python
np.unique(y)
```

```python
def plot_iris(X,y, labels = ["Iris setosa","Iris versicolor", "Iris virginica"]):
    groups = len(labels)
    Markers = ["yo", "bs", "g^", "r*"]
    plt.figure(figsize=(12, 4))
    for k in groups:
        plt.plot(X[y==k, 2], X[y==k, 3], Markers[k], label=labels[k])
    plt.xlabel("Petal length", fontsize=14)
    plt.ylabel("Petal width", fontsize=14)
    plt.legend(fontsize=12)
    plt.show()
```

```python
plot_iris(X,y, labels = ["Iris setosa","Iris versicolor", "Iris virginica"])
```

```python
from sklearn.mixture import GaussianMixture
y_pred = GaussianMixture(n_components=3, random_state=42).fit(X).predict(X)
```

```python
from scipy import stats

mapping = {}
for class_id in np.unique(y):
    mode, _ = stats.mode(y_pred[y==class_id])
    mapping[mode[0]] = class_id

mapping
```

```python
y_pred = np.array([mapping[cluster_id] for cluster_id in y_pred])
plot_iris(X,y_pred, labels = ["Cluster 1","Cluster 3","Cluster 3"])
```

```python
print(np.sum(y_pred==y))
print(np.sum(y_pred==y) / len(y_pred))
```

## Decision Boundaries - Voronoi diagram

```python
from matplotlib import cm

def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids, weights=None, circle_color='w', cross_color='r'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=35, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=2, linewidths=12,
                color=cross_color, zorder=11, alpha=1)

def plot_boundaries(clusterer, X, resolution=1000, show_centroids=True,
                             show_xlabels=True, show_ylabels=True):
    mins = X.min(axis=0) - 0.1
    maxs = X.max(axis=0) + 0.1
    xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], resolution),
                         np.linspace(mins[1], maxs[1], resolution))
    Z = clusterer.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]),cmap=cm.RdBu,alpha =0.5)
    plt.contour(Z, extent=(mins[0], maxs[0], mins[1], maxs[1]), linewidths=1, colors='k')
    plot_data(X)
    if show_centroids:
        plot_centroids(clusterer.cluster_centers_)

    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
```

```python
plt.figure(figsize=(10, 4))
plot_boundaries(kmeans, X)
plt.show()
```

# K-Means Algorithm

```python
kmeans1 = KMeans(n_clusters =5 , init ="random", n_init =1, algorithm ="full", max_iter =1, random_state =0)
kmeans2 = KMeans(n_clusters =5 , init ="random", n_init =1, algorithm ="full", max_iter =2, random_state =0)
kmeans3 = KMeans(n_clusters =5 , init ="random", n_init =1, algorithm ="full", max_iter =3, random_state =0)
kmeans1.fit(X)
kmeans2.fit(X)
kmeans3.fit(X)
```

```python
plt.figure(figsize = (12,10))
plt.subplot(321)
plot_data(X)
plot_centroids(kmeans1.cluster_centers_, circle_color ="w", cross_color ="r")
plt.ylabel("$x_2$")
plt.tick_params(labelbottom = False)
plt.title("The initially randomly centroids")

plt.subplot(322)
plot_boundaries(kmeans1, X, show_xlabels =False, show_ylabels =False)
plt.title("Labels")

plt.subplot(323)
plot_boundaries(kmeans1, X, show_centroids=False, show_xlabels=False)
plot_centroids(kmeans2.cluster_centers_)

plt.subplot(324)
plot_boundaries(kmeans2, X, show_xlabels=False, show_ylabels=False)

plt.subplot(325)
plot_boundaries(kmeans2, X, show_centroids=False)
plot_centroids(kmeans3.cluster_centers_)

plt.subplot(326)
plot_boundaries(kmeans3, X, show_ylabels=False)
```

## K-Means Variability

```python
def plot_clusterer_comparison(clusterer1, clusterer2, X,title1 =None, title2 =None):
    clusterer1.fit(X)
    clusterer2.fit(X)    
    
    plt.figure(figsize =(12,4))
    plt.subplot(121)
    plot_boundaries(clusterer1, X)
    if title1:
        plt.title(title1)
    
    plt.subplot(122)
    plot_boundaries(clusterer2, X, show_ylabels =False)
    if title2:
        plt.title(title2)
```

```python
kmeans_rd_init1 =KMeans(n_clusters=6, init ="random", n_init=1, algorithm="full", random_state=2)
kmeans_rd_init2 =KMeans(n_clusters=6, init ="random", n_init=1, algorithm="full", random_state=5)

plot_clusterer_comparison(kmeans_rd_init1, kmeans_rd_init2, X, "random state =1", "random state =2")
plt.show()
```
